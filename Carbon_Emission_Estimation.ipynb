```python
#old X and y assignment accordng to the old method

for feature in categorical_features:
    data[feature] = label_encoder.fit_transform(data[feature])

data[numerical_features] = scaler.fit_transform(data[numerical_features])
data = pd.get_dummies(data, columns=categorical_features)

X = data.loc[:, data.columns != 'CarbonEmission']
y = data['CarbonEmission']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    median_absolute_error,
    r2_score,
    explained_variance_score,
    mean_absolute_percentage_error,

)


def regression_metrics(y_true, y_pred):
    metrics = {
        "Mean Absolute Error (MAE)": mean_absolute_error(y_true, y_pred),
        "Mean Squared Error (MSE)": mean_squared_error(y_true, y_pred),
        "Root Mean Squared Error (RMSE)": np.sqrt(mean_squared_error(y_true, y_pred)),
        "Median Absolute Error": median_absolute_error(y_true, y_pred),
        "R2 Score (Coefficient of Determination)": r2_score(y_true, y_pred),
        "Explained Variance Score": explained_variance_score(y_true, y_pred),
        "Mean Absolute Percentage Error (MAPE)": mean_absolute_percentage_error(y_true, y_pred),

    }

    return metrics
```

```python
#model no. 1: backpropagation happens when the model.fit method is called
from tensorflow.keras.optimizers import Adam


model1 = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # Input layer with ReLU activation
    keras.layers.Dense(32, activation='tanh'),  # Hidden layer
    keras.layers.Dense(1)  # Output layer (linear activation for regression)
])

optimizer = Adam(learning_rate=0.001)


model1.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])


model1.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

```

```python
y_pred0 = model1.predict(X_test).flatten()

results0 = regression_metrics(y_test, y_pred0)

for metric, value in results0.items():
    print(f"{metric}: {value:.4f}")
```

```python
X_train1 = X_train.astype(np.float32).values
y_train1 = y_train.astype(np.float32).values

# Convert to TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((X_train1, y_train1)).batch(32)
loss_fn = tf.keras.losses.MeanSquaredError()



X_train_final, X_val, y_train_final, y_val = train_test_split(X_train1, y_train1, test_size=0.2)


train_dataset = tf.data.Dataset.from_tensor_slices((X_train_final, y_train_final)).batch(32)
val_dataset = tf.data.Dataset.from_tensor_slices((X_val, y_val)).batch(32)
```

```python
#model no. 2: Manually performing the forward and backward pass
# Training loop with validation

from tensorflow.keras.optimizers import Adam

epochs = 10
optimizer = Adam(learning_rate=0.001)
model1.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])
for epoch in range(epochs):
    epoch_loss = 0
    for step, (x_batch, y_batch) in enumerate(train_dataset):
        with tf.GradientTape() as tape:
            predictions = model1(x_batch, training=True)
            loss = loss_fn(y_batch, tf.squeeze(predictions))

        gradients = tape.gradient(loss, model1.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model1.trainable_variables))
        epoch_loss += loss.numpy()

    # Validation loss computation
    val_loss = 0
    for val_x, val_y in val_dataset:
        val_pred = model1(val_x, training=False)
        v_loss = loss_fn(val_y, tf.squeeze(val_pred))
        val_loss += v_loss.numpy()

    print(f"Epoch {epoch + 1}, Train Loss: {epoch_loss / (step + 1):.4f}, "
          f"Val Loss: {val_loss / len(val_dataset):.4f}")
```

```python
y_pred1 = model1.predict(X_test).flatten()

results1 = regression_metrics(y_test, y_pred1)

for metric, value in results1.items():
    print(f"{metric}: {value:.4f}")
```

```python
#model no. 2: Changing the activation function from relu to leaky relu

from tensorflow import keras
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.optimizers import Adadelta


model2 = keras.Sequential([
    Dense(128, input_shape=(X_train.shape[1],)),  # First hidden layer
    LeakyReLU(negative_slope=0.01),

    Dense(64),  # Second hidden layer
    LeakyReLU(negative_slope=0.01),

    Dense(32),  # Third hidden layer
    LeakyReLU(negative_slope=0.01),

    Dense(16),  # Fourth hidden layer
    LeakyReLU(negative_slope=0.01),

    Dense(1)  # Output layer
])

optimizer = Adadelta(learning_rate=1.0)

model2.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])

model2.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
y_pred2 = model2.predict(X_test).flatten()

results2 = regression_metrics(y_test, y_pred2)

for metric, value in results2.items():
    print(f"{metric}: {value:.4f}")
```

```python
import pandas as pd

# Combine all metrics into a DataFrame
df_metrics = pd.DataFrame({
    "Model 0": results0,
    "Model 1": results1,
    "Model 2": results2
})
```

```python
import matplotlib.pyplot as plt

# Transpose so models are rows and metrics are columns
df_metrics = df_metrics.T

# Plot
df_metrics.plot(kind='bar', figsize=(12, 6))
plt.title("Regression Model Comparison")
plt.ylabel("Metric Value")
plt.xlabel("Models")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Metrics", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
